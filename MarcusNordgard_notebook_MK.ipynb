## I3 Binning

* One way to convert a continuous variable to categorical is through binning. 
* You can bin your continuous variable using its distribution, any external knowledge, or some other logical reasons. 
* ONLY bin continuous

##Ranges taken from internet

5200-5999	Retail Trade

```python
data.loc[data['I3'] == '-'] #no missing values
```

```python
# I3 to integers
#data['I3'] = data['I3'].str.strip()
data['I3'] = pd.to_numeric(data['I3'], errors='coerce')
data['I3'].index = data['I3'].index.astype(int)
```

```python
data.info()
```

```python
#2000-3999	Manufacturing
def assign_bins(I3):
    if (I3 >= 2000) and (I3 <= 3999):       
        return(1)
    else:       
        return(0)

    
data['bins'] = data['I3'].apply(assign_bins)  # Creating a new column bins in the dataframe

data.head()
```

```python
#services and other (7000-8999)
def assign_bins(I3):  
    if (I3 >= 7000) and (I3 <= 8999):       # Creating a bin value of 1 for services
        return(1)
    else:                               # Creating a bin value of 0 for all other I3 values
        return(0)
   

data['bins'] = data['I3'].apply(assign_bins)  # Creating a new column bins in the dataframe

data.head()
```
