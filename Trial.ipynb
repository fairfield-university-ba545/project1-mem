```python
# imports
import pandas as pd
import numpy as np
import os

import pandas as pd

data = pd.read_excel('Competition1_raw_data.xlsx')
data.head()
```

```python
#drop rows that P(IPO) has dashes in
data.drop(10, inplace=True)
data.drop(86, inplace=True)
data.drop(330, inplace=True)
data.drop(400, inplace=True)
data.drop(597, inplace=True)

data.drop(215, inplace=True)
data.drop(268, inplace=True)
data.drop(490, inplace=True)
data.drop(560, inplace=True)
data.drop(614, inplace=True)

data.drop(115, inplace=True)
data.drop(123, inplace=True)
data.drop(203, inplace=True)
data.drop(217, inplace=True)
data.drop(286, inplace=True)
data.drop(357, inplace=True)
data.drop(401, inplace=True)
data.drop(410, inplace=True)
data.drop(496, inplace=True)
data.drop(503, inplace=True)
data.drop(615, inplace=True)
data.drop(675, inplace=True)

# I3 to integers
#data['I3'] = data['I3'].str.strip()
data['I3'] = pd.to_numeric(data['I3'], errors='coerce')
data['I3'].index = data['I3'].index.astype(int)
```

```python
def assign_bins(I3):
    if (I3 >= 100) and (I3 <= 999):       
        return(1)
    if (I3 >= 1000) and (I3 <= 1499):       
        return(2)
    if (I3 >= 1500) and (I3 <= 1799):       
        return(3)
    if (I3 >= 2000) and (I3 <= 3999):       
        return(4)
    if (I3 >= 4000) and (I3 <= 4999):       
        return(5)
    if (I3 >= 5000) and (I3 <= 5199):       
        return(6)
    if (I3 >= 5200) and (I3 <= 5999):       
        return(7)
    if (I3 >= 6000) and (I3 <= 6799):       
        return(8)
    if (I3 >= 7000) and (I3 <= 8999):       
        return(9)
    if (I3 >= 9100) and (I3 <= 9729):       
        return(10)
    if (I3 >= 9900) and (I3 <= 9999):       
        return(11)
    else:       
        return(0)

data['bins'] = data['I3'].apply(assign_bins)
data.head()
```

```python
#data.loc[data['T4'] == '-'] 
#data.drop(217,inplace=True)
#percent of long sentences
data["T1prime"] = data["T4"].astype(float)/data["T1"].astype(float)
#percent of real words
data["T3prime"] = data["T3"].astype(float)/data["T2"].astype(float)
#percent of long words
data["T4prime"] = data["T5"].astype(float)/data["T2"].astype(float)
#percent of positive words
data["S1prime"] = data["S1"].astype(float)/data["T2"].astype(float)
#percent of negative words
data["S2prime"] = data["S2"].astype(float)/data["T2"].astype(float)
#percent of uncertain words
data["S3prime"] = data["S2"].astype(float)/data["T2"].astype(float)
data
```

```python
# dashes rows to be dropped for dashes --> P(IPO), P(H), P(L), P(1Day), C3
indexNames1 = data[data['P(IPO)'] == '-' ].index
indexNames2 = data[data['P(H)'] == '-' ].index
indexNames3 = data[data['P(1Day)'] == '-' ].index
indexNames4 = data[data['C3'] == '-' ].index

# Delete these row indexes from dataFrame
data.drop(indexNames1 , inplace=True)
data.drop(indexNames2 , inplace=True)
data.drop(indexNames3 , inplace=True)
data.drop(indexNames4 , inplace=True)
```

```python
indexNames5 = data[data['C7'] == '-' ].index
data.drop(indexNames5 , inplace=True)
```

```python
# changing to correct type --> to continous 
cols = ['P(IPO)','P(H)','P(L)','P(1Day)','C1','C3','C4','C5','C6','C7','T1','T2','T3','T4','T5','S1','S2','S3']
data[cols] = data[cols].astype(float)
# to int 
data['C2'] = data['C2'].astype(int)
```

```python
# P(mid) calculation and add it to the column 
# P(mid) = (P(H)+P(L))/2
data['I3'] = data['I3'].astype(str)
data['C2'] = data['C2'].astype(int)
data['P(mid)'] = (data['P(H)'] - data['P(L)'])/2

# targets to be calculated
# make it equal to P(IPO) then we will just replace
data['Y1'] = data['P(IPO)']
data['Y2'] = data['P(IPO)']

# replace Y1 and change dtype to binary
data.loc[(data['Y1'] < data['P(mid)']),'Y1'] = 1
data.loc[(data['Y1'] > data['P(mid)']),'Y1'] = 0
data['Y1'] = data['Y1'].astype(int)

# replace Y2
data.loc[(data['Y2'] < data['P(1Day)']),'Y2'] = 1
data.loc[(data['Y2'] > data['P(1Day)']),'Y2'] = 0
data['Y2'] = data['Y2'].astype(int)

# Control values to be calculated
data['C3`'] = data['C3']
data['C3`'] = np.where((data['C3'] > 0),1,data['C3`'])
data['C3`'] = np.where((data['C3'] < 0),0,data['C3`'])
data['C3`'] = data['C3`'].astype(int)

# C5`
data['C5`'] = data['C5']/data['C6']

# C6`
data['C6calc'] = ((data['P(IPO)'] - data['P(mid)']) / data['P(mid)']) * 1
data['C6`'] = 0
data['C6`'] = np.where((data['P(IPO)'] > data['P(mid)']),data['C6calc'],data['C6`'])
data['C6`'] = np.where((data['P(IPO)'] < data['P(mid)']),0,data['C6`'])
data['C6`'] = np.where((data['P(mid)']==0),0,data['C6`'])
```
